```python
import numpy as np
import cv2
import time
import matplotlib.pyplot as plt
```

```python
import torch
import torchvision.models as models


def cal_num_parameters(model):
    num = 0
    for param in model.parameters():
        num += param.numel()
    return num


def visualize_segmap(img, nc=21):
    label_colors = np.array([(0, 0, 0),  # 0=background
                             # 1=aeroplane, 2=bicycle, 3=bird, 4=boat, 5=bottle
                             (128, 0, 0), (0, 128, 0), (128, 128, 0), (0, 0, 128), (128, 0, 128),
                             # 6=bus, 7=car, 8=cat, 9=chair, 10=cow
                             (0, 128, 128), (128, 128, 128), (64, 0, 0), (192, 0, 0), (64, 128, 0),
                             # 11=dining table, 12=dog, 13=horse, 14=motorbike, 15=person
                             (192, 128, 0), (64, 0, 128), (192, 0, 128), (64, 128, 128), (192, 128, 128),
                             # 16=potted plant, 17=sheep, 18=sofa, 19=train, 20=tv/monitor
                             (0, 64, 0), (128, 64, 0), (0, 192, 0), (128, 192, 0), (0, 64, 128)])

    r = np.zeros_like(img, dtype=np.uint8)
    g = np.zeros_like(img, dtype=np.uint8)
    b = np.zeros_like(img, dtype=np.uint8)
    for l in range(0, nc):
        idx = img == l
        r[idx] = label_colors[l, 0]
        g[idx] = label_colors[l, 1]
        b[idx] = label_colors[l, 2]
    rgb = np.stack([r, g, b], axis=2)

    return rgb


if __name__ == "__main__":
#     background = cv2.imread("./background.jpg")
    background = cv2.imread("./01.jpeg")
    resize_ratio = 2

    # Model
    model = models.segmentation.deeplabv3_resnet101(pretrained=True)
    print("\n=== The number of parameter is ", cal_num_parameters(model))

    cap = cv2.VideoCapture(0)
    while True:
        ret, pic = cap.read()
        pic = cv2.resize(pic, (pic.shape[1] // resize_ratio, pic.shape[0] // resize_ratio))
        background = cv2.resize(background, (pic.shape[1], pic.shape[0]))

        # Data (change the order -> normalize)
        img = np.transpose(pic[:, :, ::-1], (2, 0, 1))  # [H, W, 3]: BRG -> [H, W, 3]: RGB -> [3, H, W]

        img = torch.from_numpy((img / 255.0).astype('float32'))  # np.float == np.float64, torch.float == torch.float32
        mean = torch.Tensor([0.485, 0.456, 0.406]).reshape([3, 1, 1])
        std = torch.Tensor([0.229, 0.224, 0.225]).reshape([3, 1, 1])
        img_norm = ((img - mean) / std).unsqueeze(0)


        # Forward
        tick = time.time()
        model.eval()  # decrease computation and memory for gradient
        pred_prob = model(img_norm)["out"][0].data.cpu().numpy()
        tok = time.time() - tick
        print("\t prediction prob shape: {}".format(pred_prob.shape))
        print("\t takes: {:.1f} ms".format(tok * 1000))

        # Prediction post-process
        pred_ind = np.argmax(pred_prob, axis=0)

        pred_vis = visualize_segmap(pred_ind)
        mask = (pred_ind == 15)[:, :, np.newaxis]
        pred_out_with_ocean = (mask * pic + (1 - mask) * background).astype(np.uint8)
        pred_out_with_blur = (mask * pic + (1 - mask) * cv2.GaussianBlur(pic, (25, 25), 0)).astype(np.uint8)

        # Visualization
        plt.figure(figsize=(19.20, 10.80))
        plt.subplot(221)
        plt.imshow(pic[..., ::-1])
        plt.subplot(222)
        plt.imshow(pred_vis)
        plt.subplot(223)
        plt.imshow(pred_out_with_ocean[..., ::-1])
        plt.subplot(224)
        plt.imshow(pred_out_with_blur[..., ::-1])
        plt.show()
```
