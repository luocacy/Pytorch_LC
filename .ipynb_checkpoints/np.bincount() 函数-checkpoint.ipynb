np.bincount() 函数详解参见：[numpy.bincount 详解](https://blog.csdn.net/qq_41731861/article/details/120700267)

```python
import torch
import numpy as np
```

```python
# x 中最大值为 7，因此输出维度为 8（索引值为0->7）
x = np.array([3, 2, 1, 3, 1])
np.bincount(x)  # array([1, 3, 1, 1, 0, 0, 0, 1])  # 0 在 x 中出现了1次，1 在 x 中出现了3次......
```

```python
max(x)+1  # bicount()函数返回值维度
```

```python
w = np.array([0.3, 0.5, 0.2, 0.7, 1.])
np.bincount(x,  weights=w)
```

```python
# x = np.array([3, 2, 1, 3, 1])  # bin 的数量为4，索引值为 0->3
np.bincount(x, minlength=7)  # 指定 minlength=7，现在的索引值为0->6
# 因此，输出结果为：array([0, 2, 1, 2, 0, 0, 0])
```

```python
# 指定参数值小于原本的数量，则该参数失去作用，索引值不变
# x = np.array([3, 2, 1, 3, 1])
np.bincount(x, minlength=1)
```
